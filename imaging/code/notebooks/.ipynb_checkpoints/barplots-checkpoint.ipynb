```python
import numpy as np
import matplotlib.pyplot as plt

x = ['Day 2', 'Day 14']

# Sample data for partner
p_values_group_partner = [[[0.13953488372093023, 0.0625]], [[0.16666666666666666, 0.27419354838709675]]]
p_values_group_array_partner = np.array(p_values_group_partner)
data_partner = np.mean(p_values_group_array_partner, axis=0).flatten()

# Sample data for novel
p_value_group_novel = [[[0.046511627906976744, 0.1875]], [[0.06666666666666667, 0.04838709677419355]]]
p_value_group_array_novel = np.array(p_value_group_novel)
data_novel = np.mean(p_value_group_array_novel, axis=0).flatten()

# Width of each bar
bar_width = 0.35

# Create an array to represent the positions of the bars
x_positions = np.arange(len(x))

# Create a figure and axis
fig, ax = plt.subplots()

# Create the first set of bars
bars1 = ax.bar(x_positions - bar_width/2, data_partner, bar_width, label='Partner')

# Create the second set of bars
bars2 = ax.bar(x_positions + bar_width/2, data_novel, bar_width, label='Novel')

# Set the x-axis ticks and labels
ax.set_xticks(x_positions)
ax.set_xticklabels(x)

# Add labels and a legend
ax.set_xlabel('Sessions')
ax.set_ylabel('Percent of Significant Neurons')
ax.set_title('Percent of Significant Neurons for Contact Investigation')
ax.legend()

# Show the plot
plt.show()
```
