```python
import matplotlib.pyplot as plt
import numpy as np

ypoints1 = np.array([1, 4, 5])
ypoints2 = np.array([3, 3, 8])
ypoints3 = np.array([4, 5, 4])


plt.plot(ypoints1, c='k', linewidth = '5')
plt.plot(ypoints2, c='k', linewidth = '5')
plt.plot(ypoints3, c='r', linewidth = '5')
plt.xlabel('Day')
plt.ylabel('# of ensembles encoding social contact')
plt.xticks([0, 1, 2])
plt.show()
```
